```python
# Thao tác với các dự liệu trong dataFrame
df.item_price.dtype
```

```python
# Hàm Apply() để áp dụng bất kì hàm nào ta mún vào trong dataFrame
df['item_price']=df['item_price'].apply(lambda i:float(i.replace('$', '')))
```

```python
df.head()
```

```python
# tạo ra columns mới dựa trên dữ liệu có sẵn
df['Total_price'] = df['quantity']*df['item_price']
df.head()
```

```python
# tổng doanh thu
# doanh_thu = sum(df.Total_price)
# print(doanh_thu)
a = df['Total_price'].sum()
print(a)
```
