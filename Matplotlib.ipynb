# Sử dụng Pyplot API
## vẽ đồ thị rất là nhanh

```python
x = np.linspace(0,10,1000)
x
```

```python
plt.figure(figsize=(10,8))
plt.plot(x,np.sin(x),color ='green', lineStyle='dashed',label ='Sin')
plt.plot(x,np.cos(x),color ='red', lineStyle='dashed',label='Cos')
plt.title('Sin(x) and Cos(x)')
plt.xlabel('Biến x')
plt.ylabel('Sin(x) và Cos(x)')
plt.axis([0,4,-0.7,0.6])
# plt.axis('tight')
# plt.axis('equal')
plt.legend();
# plt.xlim([0,4])
# plt.ylim([-0.75,0.75]);
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
size = 1000*rng.rand(100)
color = rng.rand(100)

fig,ax = plt.subplots()
img1 = ax.scatter(x,y,s=size,c=color,cmap='viridis',alpha=0.4);
fig.colorbar(img1);
```

# Barchart

```python
drink_price= {'coca':12,
            'pepsi':34,
            'sprite':23}
fig,ax = plt.subplots()          
ax.bar(drink_price.keys(), drink_price.values());
ax.set(title='Bách hóa xanh',xlabel ='Sản phẩm',ylabel ='Giá');
```

```python
drink_price= {'coca':12,
            'pepsi':34,
            'sprite':23}
fig,ax = plt.subplots()          
ax.barh(list(drink_price.keys()), list(drink_price.values()));
ax.set(title='Bách hóa xanh',xlabel ='Giá',ylabel ='Sản phẩm');
```

# Histogram

```python
heght = np.random.normal(170,10,250)
heght[:10]
```

```python
plt.hist(heght,bins=50);
```

# Subplots

```python
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize= (10,8))
ax1.plot(x,x/2)
ax2.scatter(x,y)
ax3.bar(drink_price.keys(), drink_price.values())
ax4.hist(heght,bins=50);
```

```python
fig,ax = plt.subplots(nrows=2,ncols=2,figsize= (10,8))
ax[0,0].plot(x,x/2)
ax[0,1].scatter(x,y)
ax[1,0].bar(drink_price.keys(), drink_price.values())
ax[1,1].hist(heght,bins=50);
```
